```python
warnings.catch_warnings()
env = gym.make('Sudoku-v0')
env.reset()
print(env.grid)
starting_grid = env.grid
```

```python
print(f"Average Solve Time {round(sum(solve_times)/len(solve_times),4)} seconds")
```
